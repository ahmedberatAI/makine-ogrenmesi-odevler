## Özdeğer ve Özvektör Analizi

Bu çalışmada bir matrisin özdeğerleri (eigenvalues) ve özvektörleri (eigenvectors) hesaplanacaktır.  
Ayrıca bu kavramların makine öğrenmesindeki yeri kısaca incelenecektir.

```python
import numpy as np

#Matris tanımı
A= np.array([[2,1],
            [1,2]])

#eigenvalue ve eigenvector hesabı
eigenvalues, eigenvectors= np.linalg.eig(A)

#printler
print("Eigenvalues:\n")
print(eigenvalues)

print("\nEigenvectors:\n")
print(eigenvectors)

```

## Sonuç Değerlendirmesi

- Matrisin özdeğerleri ve özvektörleri `numpy.linalg.eig` fonksiyonu ile hesaplandı.
- En büyük özdeğer, matrisin en güçlü etki yaptığı yönü temsil eder.
- Özvektörler, bu etkinin gerçekleştiği doğrultuları gösterir.

## Not

Özdeğer ve özvektör kavramları, özellikle PCA gibi boyut indirgeme yöntemlerinde kullanılır.

## Özdeğerlerin Manuel Hesaplanması

Bu bölümde `numpy.linalg.eig` kullanılmadan özdeğerler, karakteristik denklem üzerinden hesaplanmıştır.

```python


# birim matris
I = np.eye(2)


a, b = A[0]
c, d = A[1]



# katsayılar
trace = a + d
determinant = a*d - b*c

print("Trace (a+d):", trace)
print("Determinant (ad-bc):", determinant)
```

```python


coefficients = [1, -trace, determinant]

eigenvalues_manual = np.roots(coefficients)

print("Manuel Eigenvalues:\n")
print(eigenvalues_manual)
```
